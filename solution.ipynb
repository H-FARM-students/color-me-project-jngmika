```python
from PIL import Image
import sys
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.colors import LogNorm
```

### Mystery 1

```python
df1 = open('data/Mystery1.data').readlines()
```

```python
for i in range(len(df1)):
    df1[i] = list(df1[i].split(', '))
    for c in df1[i].copy():
        if not c.isdigit():
            df1[i].remove(c)
```

```python
df1 = [list(map(int, row)) for row in df1]
```

```python
df1 = np.asarray(df1)
```

```python
plt.imshow(df1, cmap='viridis', norm=LogNorm())
plt.colorbar(label='My Numbers')
plt.savefig('colored1.png')

plt.show()
plt.close()
```
